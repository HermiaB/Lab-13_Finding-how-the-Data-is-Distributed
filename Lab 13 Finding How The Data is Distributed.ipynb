### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Display column names
print("Number of columns:", df.shape[1])

# Display data types 
df.info()

# Display basic summary statistics 
df.describe()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

# Fill missing Employment and RemoteWork with "Unknown"
df['Employment'].fillna('Unknown', inplace=True)
df['RemoteWork'].fillna('Unknown', inplace=True)

# Drop rows where JobSat or YearsCodePro is missing (essential for analysis)
df = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Convert YearsCodePro to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

# Calculate value counts for the 'Employment' column
employment_counts = df['Employment'].value_counts()
print("Value Counts for Employment:")
print(employment_counts)

print("\n") 

# Calculate value counts for the 'JobSat' column
jobsat_counts = df['JobSat'].value_counts()
print("Value Counts for JobSat:")
print(jobsat_counts)

print("\n")

# Calculate value counts for the 'YearsCodePro' column
yearscodepro_counts = df['YearsCodePro'].value_counts()
print("Value Counts for YearsCodePro:")
print(yearscodepro_counts)
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
!pip install pandas
!pip install matplotlib
!pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='JobSat', fill=True) 
plt.title('Kernel Density Estimate Plot of Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Density')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

# Drop missing values in the LanguageHaveWorkedWith column
df_lang = df['LanguageHaveWorkedWith'].dropna()

# Explode the semicolon-separated languages
df_lang = df_lang.str.split(';').explode()

# Strip extra spaces (if any)
df_lang = df_lang.str.strip()

# Count all language frequencies
language_counts = df_lang.value_counts()

# Define your custom list of languages
L = ['Python', 'JavaScript', 'C++', 'C#', 'Java', 'SQL', 'Go', 'Rust']

# Filter and compare only the languages in L
filtered_counts = language_counts[language_counts.index.isin(L)]

# Display the result
print("🔍 Frequency of languages in list L:")
print(filtered_counts.sort_values(ascending=False))
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

# Drop missing values for both columns
df_remote = df.dropna(subset=['RemoteWork', 'Country'])

# Optional: Limit to top 7 countries by response count for readability
top_countries = df_remote['Country'].value_counts().head(7).index
df_remote = df_remote[df_remote['Country'].isin(top_countries)]

# Create cross-tab: count of RemoteWork by Country
remote_counts = pd.crosstab(df_remote['Country'], df_remote['RemoteWork'])

# -------- Grouped Bar Chart --------
remote_counts.plot(kind='bar', stacked=False, figsize=(12, 6), colormap="Set2")
plt.title("Distribution of Remote Work Preferences by Country")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
!pip install scipy
!pip install numpy

import numpy as np
import pandas as pd
from scipy.stats import pearsonr

# Clean 'YearsCodePro' and convert to numeric
def convert_experience(val):
    if pd.isnull(val):
        return None
    elif val == "Less than 1 year":
        return 0.5
    elif val == "More than 50 years":
        return 51
    try:
        return float(val)
    except:
        return None

df['YearsCodeProNum'] = df['YearsCodePro'].apply(convert_experience)

#  Map 'JobSat' to numeric values
job_sat_mapping = {
    "Very dissatisfied": 1,
    "Dissatisfied": 2,
    "Neither satisfied nor dissatisfied": 3,
    "Satisfied": 4,
    "Very satisfied": 5
}
df['JobSatNum'] = df['JobSat'].map(job_sat_mapping)

# Step 4: Drop rows with missing values in either column
df_corr = df.dropna(subset=['YearsCodeProNum', 'JobSatNum'])

# Step 5: Check how many valid rows remain
print(f" Total responses: {len(df)}")
print(f" Valid rows for correlation: {len(df_corr)}")

# Step 6: Compute correlation only if enough data is available
if len(df_corr) >= 2:
    pearson_corr, _ = pearsonr(df_corr['YearsCodeProNum'], df_corr['JobSatNum'])
 
    print(f"\n Pearson Correlation (linear relationship): {pearson_corr:.3f}")
else:
    print("\n Not enough data to compute correlation. Showing result with simulated data.")

    # Step 7: Use dummy data if real data is insufficient
    np.random.seed(42)
    df_dummy = pd.DataFrame({
        'YearsCodeProNum': np.random.randint(0, 30, size=100),
        'JobSatNum': np.random.randint(1, 6, size=100)
    })

    pearson_corr, _ = pearsonr(df_dummy['YearsCodeProNum'], df_dummy['JobSatNum'])

    print(f"\n Pearson Correlation (Dummy Data): {pearson_corr:.3f}")
   
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here


# Drop rows with missing Employment or EdLevel
df_clean = df.dropna(subset=['Employment', 'EdLevel'])

# Cross-tabulation
crosstab = pd.crosstab(df_clean['EdLevel'], df_clean['Employment'])

# Plot a stacked bar chart
crosstab.plot(kind='bar', stacked=True, figsize=(12, 7), colormap="tab20")
plt.title("Employment Type by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title="Employment Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
